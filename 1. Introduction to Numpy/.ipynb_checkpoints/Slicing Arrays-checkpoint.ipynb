```python
#import the library and generate de 1d array from 1 to 36
import numpy as np
a1 = np.arange(1, 37)
a1
```

```python
# Use rechape function to transform the array into a 2d 6x6 array
a1=a1.reshape(6,6)
```

```python
# Obtain a 1d array of shape (6) containing the blue entries
q1= a1[:,1]
print(q1.shape)
q1
```

```python
# Obtain a two-dimensional array of shape (2,2) containing the yellow entries
q2 = a1[0:2,4:6]
print(q2.shape)
q2
```

```python
# Obtain a one-dimensional array of shape (2) containing the red entries.
q3=a1[-1,4:6]
print(q3.shape)
q3
```

```python
# Obtain a two-dimensional array of shape (2,3) containing the green entries.
q4_1=a1[2,0:5:2]
q4_2=a1[4,0:5:2]
q4 = np.array([q4_1,q4_2])
print(q4.shape)
q4
```

```python
#Alternatively the last one can be coded as:
q_a=a1[2::2, ::2]
q_a
```
