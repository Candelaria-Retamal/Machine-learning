###  Exercise: merging with different joins

Write the correct merging command to obtain each of the merged DataFrames below:

```python
from IPython.display import Image
import pandas as pd
import numpy as np

Image(filename="Exercise 4.png")
```

```python
left = pd.DataFrame(
    {"key1": ["a", "b", "c"], "key2": ["A", "B", "C"], "lval": [0, 1, 2]}
)
right = pd.DataFrame(
    {"key1": ["a", "b", "c"], "key2": ["A", "D", "C"], "rval": [3, 4, 6]}
)
left
```

```python
right
```

```python
# A. Only identical rows are added using key1 and key2 as the shared columns. No NaN and no repeated name of cols.
pd.merge(left,right)
```

```python
# B. All of the data is concatenated, Yes Nan values. No repeated name of cols or rows.
pd.merge(left,right, on=["key1","key2"],how="outer")
```

```python
# C. Data contained in BOTH key1 and key2 are incorporated + data only in right. Yes NaN values. No repeated col names or row names
pd.merge(left,right,how="right")
```

```python
# D. Data is merged using key1 as the common column. No NaN values. No repeated names of cols and rows
pd.merge(left,right,on="key1",how="right")
```

#### Exercise: Grouping and filtering

Using df2 your task is to now filter this data according to the following rule:

a group is kept if and only if all the values of the group are greater than 50 in both score1 and score2.
You can find the solution video below. As usual, first give the exercise an honest try.

```python
import pandas as pd
import numpy as np

raw_data = {'team': ['Ten Snakes', 'Ten Snakes', 'Ten Snakes', 'Ten Snakes', 
                     'Nine Monkeys', 'Nine Monkeys', 'Nine Monkeys', 'Nine Monkeys', 
                     'Eight Eagles', 'Eight Eagles'], 
        'rank': ['1st', '1st', '2nd', '2nd', '1st', '1st', '2nd', '2nd','1st', '2nd'], 
        'name': ['James', 'Allen', 'Matthew', 'James', 'Devon', 'Sam', 'Justin', 'Sam', 'Paul', 'Ross'], 
        'score1': [16,35,55,29,2,61,68,41,94,18],
        'score2': [81,65,54,44,28,93,2,5,53,99]}
df = pd.DataFrame(raw_data, columns = ['team', 'rank', 'name', 'score1', 'score2'])
df2 = df.set_index(["team", "rank"])
grouped2 = df2.groupby(level=['team', 'rank'])
list(grouped2)
```

```python
def f(x):
    m = x.min()
    return (m.score1 > 50) & (m.score2 > 50)
grouped2.filter(f)
```
