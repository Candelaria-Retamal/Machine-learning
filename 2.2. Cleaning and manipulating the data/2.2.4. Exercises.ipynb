#### Removing duplicates
Import the following dataset:

```python
import pandas as pd

df = pd.read_csv("http://bit.ly/chiporders", sep="\t")
```

Perform the following steps:

* Find the number of duplicated rows.
* Find the number of duplicated values in each column.
* Find the number of duplicated rows with respect to the columns 'order_id', 'item_name' and 'item_price'.
* Remove all occurrences of the duplicated rows with respect to the columns 'order_id', 'item_name' and 'item_price'.

#### Solution
Find the number of duplicated rows.

```python
df.duplicated().sum()
```

Find the number of duplicated values in each column.

```python
df.info()
```

```python
column_names=df.columns
for x in column_names:
    print(str(df.duplicated([x]).sum())+str(" in column ")+ x)
```

Find the number of duplicated rows with respect to the columns 'order_id', 'item_name' and 'item_price'.

```python
# In these three columns all of the entries that are the same if True (already appeared in previous values)
df.duplicated(['order_id', 'item_name', 'item_price']).sum()
```

Remove all occurrences of the duplicated rows with respect to the columns 'order_id', 'item_name' and 'item_price'.

```python
df.drop_duplicates(['order_id', 'item_name', 'item_price'], keep=False) #with keep=False we maintain only the arguments that are false to duplication
```

2) Calories can be converted to kilojoules using the following formula: 1 calorie = 4.184 kilojoules. Find the amount of Kilojoules per serving for each cereal. Store the results in a new column.

3) For each numerical column, find the cereal with the maximum value per column.

```python
# Import libraries
import pandas as pd
import numpy as np

# Load data
df = pd.read_csv("c2_cereal.csv")

# Print data shape
print('Data shape: ',df.shape)
```

```python
df["Manufacturer"] = df.mfr.map({"A": "American Home Food Products", "G": "General Mills", "K": "Kelloggs", "N": "Nabisco", "P": "Post", "Q": "Quaker Oats", "G": "Ralston Purina"})
df
```

```python
def kj(x):
    y = x*4.184
    return(y)

df["KJ"]=df.calories.apply(kj)
df
```

```python
df.columns.to_list()
df_num = df.loc[:, ['calories','protein','fat','sodium','fiber','carbo','sugars','potass','vitamins','shelf','weight','cups','rating', 'KJ']]
df_num.apply(max, axis=0)
```
